# Plotting

```python
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
mpg.head(10)
```

```python
mpg.tail(10)
```

```python
mpg.dtypes
```

```python
mpg.shape
```

```python
mpg.info()
```

```python
# to visualise a column's data, we can call the method '.plot()'
mpg['mpg'].plot()
```

```python
# to generate a histogram
mpg['mpg'].plot(kind='hist')
```

```python
# to generate a box plot
mpg['mpg'].plot(kind='box')
```

```python
# to get the origin of the cars we can use the .value_counts() method
mpg['origin'].value_counts()
```

```python
# we can visualise this data using an horizontal barchart
mpg['origin'].value_counts().plot(kind = 'barh')
```

```python
# visualise the data in a pie chart
mpg['origin'].value_counts().plot(kind = 'pie')
```

```python
# to create a scatter chart of the cars' weight and mpg's
mpg.plot(x = 'weight', y = 'mpg', kind = 'scatter')
```
